```
def sentiment_analizer(text):
  res = sonar.ping(text=text)
  #print(res)
  class1 = res['classes'][0]['class_name']
  class2 = res['classes'][1]['class_name']
  conf1 = res['classes'][0]['confidence']
  conf2 = res['classes'][1]['confidence']

  return pd.Series([conf1, conf2]) 
```

```
sentiment_analizer('Happy')
```

```
data[['asari_negative','asari_positive']] = data['review'].apply(sentiment_analizer)
```

```
def BERT_sentiment_analizer(text):
  res = sentiment_analyzer(text)
  #print(res)
  label = res[0]['label']
  score = res[0]['score']

  return pd.Series([label, score]) 
```

```
data[['BERT_result_label','BERT_result_score']] = data['review'].apply(BERT_sentiment_analizer)
```
